### Prothom Alo Step 1: Invalid news filtering

```python
def get_date_from_text(text):
    if text[:7] == 'আপডেট: ':
        text = text[14:]
    else:
        text = text[7:]
    year = replace_bangla_digits(text[-4:])
    day = replace_bangla_digits(text[-8:-6])
    month = get_month_from_bangla(text[:-9])
    return day + '-' + month + '-' + year

def get_month_from_bangla(text):
    month_dict = {'জানুয়ারি': '01', 'ফেব্রুয়ারি': '02', 'ফেব্রুয়ারি':'02', 'মার্চ': '03', 'এপ্রিল': '04', 'মে': '05', 'জুন': '06',
                  'জুলাই': '07', 'আগস্ট': '08', 'সেপ্টেম্বর': '09', 'অক্টোবর': '10', 'নভেম্বর': '11', 'ডিসেম্বর': '12'}
    if text in month_dict:
        return month_dict[text]
    print('Error: month not found from text ' + text)    
    return '00'

def replace_bangla_digits(text):
    text = text.replace('০', '0')
    text = text.replace('১', '1')
    text = text.replace('২', '2')
    text = text.replace('৩', '3')
    text = text.replace('৪', '4')
    text = text.replace('৫', '5')
    text = text.replace('৬', '6')
    text = text.replace('৭', '7')
    text = text.replace('৮', '8')
    text = text.replace('৯', '9')
    return text

def filter_prothom_alo(data_handle, filter_handle, output_handle):
    with open(filter_handle, 'r', encoding='utf-8-sig') as invalid_file:
        invalid_headlines = [x.strip().replace("ৌ", "ৌ").replace("য়", "য়").replace("ড়", "ড়") for x in invalid_file.readlines()]
        #print(invalid_headlines)
    with open(data_handle, 'r',  encoding='utf8') as data_raw:
        data_raw.readline()
        articles = data_raw.read().split('#####\n#####\n')
        with open(output_handle, 'w', encoding='utf8') as out:
            for i in range(len(articles)):
                bits = articles[i][2:-37].split('--\n--')
                if bits[4] in invalid_headlines:
                    continue
                out.write(bits[0] + '\n')
                out.write(bits[2] + '\n')
                out.write(bits[3] + '\n')
                out.write(bits[4] + '\n')
                out.write(get_date_from_text(bits[5]) + '\n')
                out.write(bits[6].replace("ৌ", "ৌ").replace("য়", "য়").replace("ড়", "ড়"))
                if i < len(articles) - 1:
                    out.write('\n#######\n')

data_handle = 'D:/Sachetan/Newscrawl/raw/prothom-alo_mugging.txt'
filter_handle = 'D:/Sachetan/Newscrawl/prothom-alo/invalid_headlines_mugging.txt'
output_handle = 'D:/Sachetan/Newscrawl/prothom-alo/prothom_alo_mugging_step-1.txt'
filter_prothom_alo(data_handle, filter_handle, output_handle)
data_handle = 'D:/Sachetan/Newscrawl/raw/prothom-alo_road-accident.txt'
filter_handle = 'D:/Sachetan/Newscrawl/prothom-alo/invalid_headlines_road-accident.txt'
output_handle = 'D:/Sachetan/Newscrawl/prothom-alo/prothom_alo_road-accident_step-1.txt'
filter_prothom_alo(data_handle, filter_handle, output_handle)
data_handle = 'D:/Sachetan/Newscrawl/raw/prothom-alo_robbery.txt'
filter_handle = 'D:/Sachetan/Newscrawl/prothom-alo/invalid_headlines_robbery.txt'
output_handle = 'D:/Sachetan/Newscrawl/prothom-alo/prothom_alo_robbery_step-1.txt'
filter_prothom_alo(data_handle, filter_handle, output_handle)
data_handle = 'D:/Sachetan/Newscrawl/raw/prothom-alo_sexual-harassment.txt'
filter_handle = 'D:/Sachetan/Newscrawl/prothom-alo/invalid_headlines_sexual-harassment.txt'
output_handle = 'D:/Sachetan/Newscrawl/prothom-alo/prothom_alo_sexual-harassment_step-1.txt'
filter_prothom_alo(data_handle, filter_handle, output_handle)
data_handle = 'D:/Sachetan/Newscrawl/raw/prothom-alo_theft.txt'
filter_handle = 'D:/Sachetan/Newscrawl/prothom-alo/invalid_headlines_theft.txt'
output_handle = 'D:/Sachetan/Newscrawl/prothom-alo/prothom_alo_theft_step-1.txt'
filter_prothom_alo(data_handle, filter_handle, output_handle)
```

### Upazilla dictionary builder

```python
class Upazilla:
    pass

def build_upazila_dict(upazilla_file_handle = '201314upazilawithzila.txt'):
    upazilla_dict = {}
    with open(upazilla_file_handle, 'r', encoding = 'utf8') as upazilla_raw:
        for line in upazilla_raw.readlines():
            words = line.strip().split(',')
            upazilla = Upazilla()
            upazilla.upazilla_name = words[0].replace("ৌ", "ৌ").replace("য়", "য়").replace("ড়", "ড়")
            upazilla.upazilla_name_bn = words[1].replace("ৌ", "ৌ").replace("য়", "য়").replace("ড়", "ড়")
            upazilla.zilla_name = words[2].replace("ৌ", "ৌ").replace("য়", "য়").replace("ড়", "ড়")
            upazilla.zilla_name_bn = words[3].replace("ৌ", "ৌ").replace("য়", "য়").replace("ড়", "ড়")
            upazilla_dict[upazilla.upazilla_name_bn] = upazilla
    return upazilla_dict
```

### Prothom Alo Step 2: adding location tag

```python
def process_tagging(input_handle, output_handle):
    with open(input_handle, 'r', encoding = 'utf-8') as file:
        articles = file.read().split('#######\n')
        for article in articles:
            lines = article.strip().split('\n')
            print(lines[2])
            print(lines[3])
            total_locations_found = find_locations_from_text(lines[5], lines[2].split())
            for location in total_locations_found:
                print(location)
            print('----------------------')
            
def find_locations_from_text(text, keywords):
    sentences = text.split('।')
    total_locations = set()
    for i in range(len(sentences)):
        sentence = sentences[i]
        for keyword in keywords:
            if keyword in sentence:
                # if the keyword and upazilla is in one sentence, extract upazilla and finish processing keyword occurance
                tagged_upazillas = find_upazilla_from_sentence(sentence)
                if tagged_upazillas: # check if set not empty
                    total_locations = total_locations | tagged_upazillas
                    continue
                # if keyword and upazilla not in same sentence, look at neighboring sentences
                else:
                    j, k = i-1, i+1
                    # look behind
                    while j >=0 and i-j < 4:
                        previous = sentences[j]
                        another_keyword_found = False
                        for keyword in keywords:
                            if keyword in previous:
                                another_keyword_found = True
                        if another_keyword_found:
                            break
                        total_locations = total_locations | find_upazilla_from_sentence(previous)
                        j -= 1
                    while k < len(sentences) and k-i < 4:
                        next = sentences[k]
                        another_keyword_found = False
                        for keyword in keywords:
                            if keyword in next:
                                another_keyword_found = True
                        if another_keyword_found:
                            break
                        total_locations = total_locations | find_upazilla_from_sentence(next)
                        k += 1
    return total_locations
                        
                        
                
def find_upazilla_from_sentence(sentence):
    upazilla_file_handle = 'D:/Sachetan/Newscrawl/201314upazilawithzila.txt'
    upazilla_dict = build_upazila_dict(upazilla_file_handle)
    found_upazillas = set()
    for upazilla_name_bn in upazilla_dict.keys():
        if upazilla_name_bn in sentence:
            found_upazillas.add(upazilla_dict[upazilla_name_bn].upazilla_name)
    return found_upazillas
        
input_handle = 'D:/Sachetan/Newscrawl/prothom-alo/prothom_alo_mugging_step-1.txt'
output_handle = 'D:/Sachetan/Newscrawl/prothom-alo/prothom_alo_test_step-2.txt'
process_tagging(input_handle, output_handle)
```
